```python
#Importing libraries
import numpy as np
import statistics as st
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
#Data
data = np.array([25,14,58,69,32,14,45,22,33,65,14,74,85,36])
data
```

```python
#Creating boxplot
plt.figure(figsize = (16,9))
plt.boxplot(data);
```

# Range or IQR

```python
#Computing range manually
quantile_range = np.percentile(data,75)-np.percentile(data,25)
quantile_range
```

```python
#Computing range
stats.iqr(data)
```

# Quantile

```python
#Computing 1st quantile
np.percentile(data,25)
```

```python
#Computing 1st quantile
np.quantile(data,0.25)
```

```python
#Computing 2nd quantile
np.percentile(data,50)
```

```python
#Computing 2nd quantile
np.quantile(data,0.50)
```

```python
#Computing 3rd quantile
np.percentile(data,75)
```

```python
#Computing 3rd quantile
np.quantile(data,0.75)
```

```python
#Creating boxplot
plt.figure(figsize = (16,9))
plt.boxplot(data);
plt.axhline(22.75);
plt.axhline(63.25);
plt.axhline(34.5);
plt.axhline(40.5);
```

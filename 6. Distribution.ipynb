# Normal Distribution 

```python
#Loading libraries
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, t
```

```python
#Creating 2 by 3 random matrix
x = random.normal(size=(2, 3))
x
```

```python
#Creating random normal distribution where mean is 1 and standard deviation is 2
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
#Creating normal distribution plot
sns.distplot(random.normal(size=1000,loc=0, scale=1), hist=False);
plt.show()
```

# Uniform Distribution

```python
#Creating normal distribution plot
sns.distplot(random.uniform(size=1000), hist=False);
plt.show()
```

# Exponential Distribution

```python
#Creating normal distribution plot
sns.distplot(random.exponential(size=1000), hist=False);
plt.show()
```

# Binomial Distribution

```python
#Creating normal distribution plot
sns.distplot(random.binomial(n=10, p=0.5,size=1000), hist=True, kde=False);
plt.show()
```

# Poisson Distribution

```python
#Creating normal distribution plot
sns.distplot(random.poisson(lam=2,size=1000), hist=True, kde=False);
plt.show()
```

# Logistic Distribution

```python
#Creating normal distribution plot
sns.distplot(random.logistic(size=1000), hist=False);
plt.show()
```

# Chi Square Distribution

```python
#Creating normal distribution plot
sns.distplot(random.chisquare(df=1,size=1000), hist=False);
plt.show()
```

# Gamma Distribution

```python
#Creating normal distribution plot
sns.distplot(random.gamma(shape=1,scale=1,size=1000), hist=False);
plt.show()
```

# Bernoulli Distribution

```python
#Creating normal distribution plot
sns.distplot(bernoulli.rvs(size=1000,p=0.6), hist=True, kde=True);
plt.show()
```

# T Distribution

```python
#Creating normal distribution plot
numargs = t.numargs
a,b= 4.32, 3.18
rv = t(a,b)
rv
quantile = np.arange(0.01, 1, 0.1) 
# Random Variates 
R = t.rvs(a, b) 
print ("Random Variates : \n", R) 
  
# PDF 
R = t.pdf(a, b, quantile) 
print ("\nProbability Distribution : \n", R) 
```

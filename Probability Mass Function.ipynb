# Probability Mass Function - Countinues Random Variable

```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating list
list1 = [1,1,6,3,2,3,4,2,3,5,5,6,5,5,5,5,5,2,2,3,3,5,2,2,5,6,2,2,2,3,6,6,2,4,3,2,3]
list1
```

```python
#List converting into dataframe
df = pd.DataFrame(list1)
df
```

```python
#Computing value counts
df1 = df[0].value_counts()
df1
```

```python
#Sum of values
sum1 = len(df)
sum1
```

```python
#Converting value counts into dataframe
df2 = pd.DataFrame(df1)
df2
```

```python
#Changing column name 
df2 = df2.rename(columns={0:'item_counts'})
df2
```

```python
#Creating item column of index number
df2['item'] = df2.index
df2
```

```python
#Computing probability
df2['probability'] = df2['item_counts']/sum1
df2
```

```python
#Ploting 
plt.bar(df2['item'],df2['probability'],color='r')
plt.show()
```

# Probability Mass Function - Discrete Random Variable

```python
#Creating data
d = pd.DataFrame()
d['color'] = ['Red','Green','Yellow','Black','Orange']
d['count'] = [5,8,6,2,9]
d
```

```python
#Computing probability
d['pmf'] = d['count']/d['count'].sum()
d
```

```python
#Ploting
plt.bar(d['color'],d['pmf'])
plt.show()
```

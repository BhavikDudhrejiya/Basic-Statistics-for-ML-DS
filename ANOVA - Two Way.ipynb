Hypothesis:
- H0 - variables are not statistically significant for impacting height
- H1 - variables are statistically significant for impacting height

Criteria:
 - p-value<=0.05 - Reject the null hypothesis H0 - statistically significant for impacting heigh
 - p_value>0.05 - Fail to reject the null hypothesis H0 - not statistically significant for impacting height

```python
#Importing libraries for data manipulation
import pandas as pd
import numpy as np

#Importing libraries for ANOVA
from statsmodels.stats.anova import anova_lm

#Importing libraries for OLS
from statsmodels.formula.api import ols
```

```python
#Creating data
df = pd.DataFrame({'water': np.repeat(['daily', 'weekly'], 15),
                   'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
                   'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5,
                              6, 6, 7, 8, 7, 3, 4, 4, 4, 5,
                              4, 4, 4, 4, 4, 5, 6, 6, 7, 8]})
```

```python
#Checking data
df
```

```python
df.keys()
```

```python
#Performing OLS
model = ols('height ~ sun + water', data=df).fit()
model.summary()
```

```python
#Two way anova
anova_lm(model, typ = 2)
```

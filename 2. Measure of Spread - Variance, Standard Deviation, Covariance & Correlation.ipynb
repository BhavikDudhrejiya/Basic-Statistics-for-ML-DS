```python
#Importing libraries
import statistics as st
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Creating data
data = np.array([25,14,58,65,32,74,15,26,65,47,58,96,25,47])
data
```

```python
#Computing mean
st.mean(data)
```

```python
#Computing variance
st.variance(data)
```

```python
#Computing manually
error = data-st.mean(data)
error2 = error**2
ss = sum(error2)
variance = ss/(len(data)-1)
variance
```

```python
#Computing standard deviation
st.stdev(data)
```

```python
#Computing manually
error = data-st.mean(data)
error2 = error**2
ss = sum(error2)
variance = ss/(len(data)-1)
standard_deviation = st.sqrt(variance)
standard_deviation
```

```python
#Computing relative standard deviation
mean = st.mean(data)
std = st.stdev(data)
rel_std = std/mean*100
mean,std,rel_std
```

```python
#Creating data
data1 = np.array([25,14,58,65,32,74,15,26,65,47,58,96,25,47])
data2 = np.array([25,45,21,65,98,74,10,23,20,25,85,45,44,21])
```

```python
#Creating plot
plt.scatter(data1, data2);
```

```python
#Creating boxplot
plt.subplot(1,2,1)
plt.boxplot(data1);
plt.subplot(1,2,2)
plt.boxplot(data2);
```

```python
#Computing covariance
error1 = data1 - st.mean(data1)
error2 = data2 - st.mean(data2)
sum_e1e2 = sum(error1*error2)
n = len(data1)
covarinace = sum_e1e2/n
covarinace
```

```python
#Computing correlation coefficient through pandas
df = pd.DataFrame()
df['x'] = data1
df['y'] = data2

df.corr()
```

```python
#Computing correlation through numpy
np.corrcoef(data1,data2)
```
